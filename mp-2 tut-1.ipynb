```python
import pyatom.lp as lp 
import pyatom.grb_solver as grb 

model = lp.Model()
x = model.dvar()
y = model.dvar()

model.max(3*x + 4*y)
model.st(2.5*x + y <=20)
model.st(3*x + 3*y <=30)
model.st(x + 2*y <=16)
model.st(x <= 3)
model.st(abs(y) <= 2)

model.solve(grb)
```

```python
print(model.get())
print(x.get())
print(y.get())
```

```python
model.do_math()
```

```python
##### 2 problem

from rsome import ro
from rsome import grb_solver as grb

model = ro.Model('LP model')
x = model.dvar()
y = model.dvar()

model.max(3*x + 4*y)
model.st(2.5*x + y <=20)
model.st(3*x + 3*y <=30)
model.st(x + 2*y <=16)
model.st(x <= 3)
model.st(abs(y) <= 2)

model.solve(grb)
```

```python
print('x:', x.get())
print('y:', y.get())
print('objective:', model.get())
```
